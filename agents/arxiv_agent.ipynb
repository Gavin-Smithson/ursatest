```python
from langchain import hub
from langchain.agents import AgentExecutor, create_react_agent, load_tools
from langchain_openai import ChatOpenAI
from langchain_community.utilities.arxiv import ArxivAPIWrapper
from langchain_community.tools      import DuckDuckGoSearchResults
from langchain_community.tools.arxiv.tool import ArxivQueryRun
import os

os.environ['OPENAI_API_KEY'] = 'xxx'

llm = ChatOpenAI(model       = "o3-mini",
                 max_tokens  = 10000,
                 timeout     = None,
                 max_retries = 2)

arxiv_wrapper = ArxivAPIWrapper(top_k_results = 10)
arxiv_tool = ArxivQueryRun(api_wrapper=arxiv_wrapper)
search_tool = DuckDuckGoSearchResults(output_format="json", num_results=10)

tools = [arxiv_tool,search_tool]


prompt = hub.pull("hwchase17/react")

agent = create_react_agent(llm, tools, prompt)

agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

```

```python
agent_executor.invoke(
    {
        "input": "Check if a neutron star radius of 22 km is consistent with what is reported in the arxiv literature.",
    }
)
```
